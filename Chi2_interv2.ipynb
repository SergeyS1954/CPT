```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2
s_2 = []
l_0 = []
l_1 = []
for st in range (5, 55, 1):
    seed = np.random.seed(12345)
    test = np.random.normal(5,1,20*st)
    ss = np.var(test)
    s_2.append(ss)
    ll = chi2.interval(0.975,st-1)[0]
    l_0.append(ss*st/ll)
    rr = chi2.interval(0.975,st-1)[1]
    l_1.append(ss*st/rr)
#print(l_0)
s_2 = np.array(s_2)
l2_0 = np.array(l_0)
l2_1 = np.array(l_1)
#print(l2_1)
yl = s_2 - l2_0
yu = s_2 + l2_1
st = np.arange(5,55,1)
plt.plot(st,s_2[st-5],label = 'среднее')
plt.plot(st,yu[st-5],label = 'верхняя граница')
plt.plot(st,yl[st-5 ],label = 'нижняя граница')
plt.legend()
#plt.savefig('chi2_interv2.jpg', dpi = 300)
plt.show()
```
