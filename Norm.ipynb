```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
x = np.linspace(2., 8., 100)
ax.plot(x, norm.pdf(x, 5, 1), 'r-', label='$\sigma$ = 1')
ax.plot(x, norm.pdf(x, 5, 1/2), 'b-',  label='$\sigma$ = 1/2')
ax.plot(x, norm.pdf(x, 5, 1/5), 'g-', label='$\sigma$ = 1/5')
plt.legend()
plt.savefig('norm.jpg', dpi = 300)
plt.show()
```
