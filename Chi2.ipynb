```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
x = np.linspace(1., 25., 100)
ax.plot(x, chi2.pdf(x, 5), 'r-', label='$n $ = 2')
ax.plot(x, chi2.pdf(x, 10), 'b-',  label='$n$ = 5')
ax.plot(x, chi2.pdf(x, 20), 'g-', label='$n$ = 10')
plt.legend()
#plt.savefig('chi2.jpg', dpi = 300)
plt.show()
```
